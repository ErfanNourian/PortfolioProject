```python
# import the packages

import pandas as pa
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import matplotlib
%matplotlib inline
```

```python
# Read the data

df = pa.read_csv('movies.csv')
```

```python
df
```

```python
# Populate nulls with 0

df['votes'] = df['votes'].fillna(0)
df['budget'] = df['budget'].fillna(0)
df['gross'] = df['gross'].fillna(0)
df['runtime'] = df['runtime'].fillna(0)
```

```python
# Change types

df['votes'] = df['votes'].astype('int64')
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
```

```python
# See all nulls in each columns

df.isnull().sum()
```

```python
# Boxplot by gross

df.boxplot(column=['gross'])
```

```python
# Sorting by gross 

df.sort_values(by=['gross'],ascending=False)
```

```python
# Regplot with gross and budget

sns.regplot(data=df, x='gross', y='budget', scatter_kws={'s':2}, line_kws={"color": "purple"})
```

```python
# Regplot with gross and score

sns.regplot(data=df, x='score', y='gross', scatter_kws={'s':2}, line_kws={"color": "purple"})
```

```python
# Pearson correlation method

df.corr(method='pearson')
```

```python
# Kendall correlation method

df.corr(method='kendall')
```

```python
# Spearman correlation method

df.corr(method='spearman')
```

```python
# Correlation matrix for numeric features

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix for numeric features')

plt.xlabel('Movie features')

plt.ylabel('Movie features')
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr()

figure(figsize=(15,10))

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix for numeric features')

plt.xlabel('Movie features')

plt.ylabel('Movie features')
```

```python
# Correlation between pairs

correlation_mat = df.apply(lambda x : x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack().sort_values(kind = 'quicksort')

corr_pairs
```

```python
# Finding the strong correlations

strong_pairs = corr_pairs[abs(corr_pairs)>0.5] 
strong_pairs
```

```python
# Looking at the top 10 compaies by gross revenue

company_gross_sum = df.groupby('company')[['gross']].sum()

company_gross_sum_sorted = company_gross_sum.sort_values('gross', ascending=False)[:10]

company_gross_sum_sorted
```
